### Relojista

```python

print("BIENVENIDO AL SISTEMA EXPERTO DE RELOJERÍA")
print("Responda las preguntas para identificar la categoría de su pieza.")
print("-------------------------------------------------------------")

movimiento = input("¿El movimiento es 'mecanico' o de 'cuarzo'?: ").lower()

if movimiento == "mecanico":
    carga = input("¿Es de carga 'automatica' (rotor) o 'manual' (cuerda)?: ").lower()

    if carga == "automatica":
        reserva = int(input("¿Cuántas horas de reserva de marcha tiene? (ej: 40): "))

        if reserva >= 70:
            print("IA: Identificado como Reloj Automático de Alta Regalía (Long Reserve).")
        elif reserva >= 38 and reserva < 70:
            print("IA: Identificado como Reloj Automático de Manufactura Estándar.")
        else:
            print("IA: Clasificado como Movimiento Automático Vintage o de entrada.")

    elif carga == "manual":
        estilo = input("¿Es de estilo 'esqueleto' (maquinaria visible)? (si/no): ").lower()
        if estilo == "si":
            print("IA: Identificado como Reloj Mecánico Manual de Exhibición (Squelette).")
        else:
            print("IA: Identificado como Reloj Mecánico Manual Clásico.")
    else:
        print("Error: El tipo de carga no está en la base de conocimientos.")

elif movimiento == "cuarzo":

    pantalla = input("¿La visualización es 'analoga' (manecillas) o 'digital'?: ").lower()

    if pantalla == "digital":

        es_inteligente = input("¿Tiene sensores de salud o GPS? (si/no): ").lower()

        if es_inteligente == "si":
            print("IA: Clasificado como Smartwatch con IA embebida (TinyML).")
        else:
            print("IA: Clasificado como Reloj Digital Electrónico Estándar.")

    elif pantalla == "analoga":
        complicacion = input("¿Tiene función de cronógrafo (cronómetro)? (si/no): ").lower()
        if complicacion == "si":
            print("IA: Clasificado como Reloj de Cuarzo Analógico Deportivo (Cronógrafo).")
        else:
            print("IA: Clasificado como Reloj de Cuarzo Analógico de Vestir.")
    else:
        print("IA: Error. Tipo de pantalla no reconocido por el sistema.")

else:
    print("IA: Error Crítico. El movimiento ingresado está fuera de mi dominio experto.")

print("-------------------------------------------------------------")
print("Simulación de diagnóstico finalizada.")
print("Programa Realizado por Zavala Medina Alexia Paola NL= 59")
```
